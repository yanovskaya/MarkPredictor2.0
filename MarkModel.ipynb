```python
import pandas
document = pandas.read_csv('Marks-Dataset.csv')
Inputs = document[['Visits', 'Homework', 'Test']].values
Outputs = document['Exam'].values
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
input_layer = Dense(2, activation='sigmoid', input_shape=(3,), use_bias = False)
hidden_layer = Dense(1, activation='linear', use_bias = False)
model.add(input_layer)
model.add(hidden_layer)
```

```python
model.summary()
```

```python
model.compile(loss="mean_squared_error", optimizer=Adam(lr=0.8))
model.fit(Inputs, Outputs, epochs=100)
```

```python
model.get_weights()
```

```python
import coremltools
coreml_model = coremltools.converters.keras.convert(model, 
                                                    input_names='matrixVisitsHomeworkTest', 
                                                    output_names='examMark')
```

```python
coreml_model.author = 'Elena Yanovskaya'
coreml_model.short_description = 'Predicts the exam mark for neural networks.'
```

```python
coreml_model.save('MarkModel.mlmodel')
```
